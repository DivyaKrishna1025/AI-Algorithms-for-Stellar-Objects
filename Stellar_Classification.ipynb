```python
import matplotlib.pyplot as pnt
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/balanced_star_classification.csv')
```

```python
df.head() 
```

```python
df['class'].value_counts()
```

```python
df.corr(method='pearson')
```

```python
# empty data frame
balanced_df = pd.DataFrame(columns = df.columns)
```

```python
label_data = df['class'].value_counts()
```

```python
for label_categories in list(label_data.index):
  sample_label = df.loc[df['class'] == label_categories]
  length = len(sample_label)
  length = min(length,18961)
  balanced_df = balanced_df.append(sample_label.iloc[0:length])
```

```python
len(balanced_df)
```

```python
balanced_df = balanced_df.sample(frac=1, random_state=42)
```

```python
balanced_df.to_csv('/content/balanced_star_classification.csv', index=False)
```

```python
barplot = balanced_df['class'].value_counts()
```

```python
galaxy = list(barplot.index)
```

```python
values = list(barplot.values)
```

```python
pnt.bar(galaxy,values,width=0.4,color='maroon')
pnt.xlabel('Class')
pnt.ylabel('Samples of class')
pnt.title('Balanced Dataset')
```
